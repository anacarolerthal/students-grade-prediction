```python
from pandas_profiling import ProfileReport
import pandas as pd

student = pd.read_csv('student-mat.csv')
```

```python
profile = ProfileReport(student, title="Students Grading Report")
# visualização do relatório
profile
```

```python
profile.to_file("profile.html")
```

```python
%matplotlib inline
# plot higher vs failures
import seaborn as sns
import matplotlib.pyplot as plt

fig = plt.figure()
sns.catplot(x="failures", y="G3", data=student, kind="violin")
fig.show()
```

```python

fig = plt.figure()
sns.catplot(x="failures", y="G2", data=student, kind="violin")
fig.show()
```

```python
fig = plt.figure()
sns.catplot(x="failures", y="G1", data=student, kind="violin")
fig.show()
```

```python
# GM = (G1 + G2 + G3) / 3
student['GM'] = (student['G1'] + student['G2'] + student['G3']) / 3

fig = plt.figure()
sns.catplot(x="failures", y="GM", data=student, kind="violin")
fig.show()
```

```python
# plot with suplots (G1, G2, G3)
fig, axes = plt.subplots(1, 3, figsize=(9, 5), sharey=True)
sns.violinplot(x="failures", y="G1", data=student, ax=axes[0])
sns.violinplot(x="failures", y="G2", data=student, ax=axes[1])
sns.violinplot(x="failures", y="G3", data=student, ax=axes[2])
```
